```python
import imageio.v3 as iio
from PIL import Image
import numpy as np

# List of uploaded images
filenames = ['1.jpg', '2.jpg','3.jpg','4.jpg','5.jpg','6.jpg','7.jpg','8.jpg','9.jpg','10.jpg','11.jpg']
images = []

# Read images one by one and resize them
target_size = (500, 500)  # Define a target size for resizing (width, height) - you can change this
for filename in filenames:
    img = Image.open(filename)
    img = img.resize(target_size)
    # Convert image to numpy array and ensure consistent color channels
    img = np.array(img)
    if img.ndim == 2: # grayscale
        img = np.stack([img]*3, axis=-1) # convert to RGB
    images.append(img)


# Save GIF
iio.imwrite('team.gif', images, duration=200, loop=0)
```
